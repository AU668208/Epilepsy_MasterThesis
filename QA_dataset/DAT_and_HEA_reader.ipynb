```python
# Dataloader for DAT and HEA files

import os
import numpy as np
import pandas as pd
from scipy.signal import welch
import matplotlib.pyplot as plt

dat_path = r"C:\Users\kvjkv\Downloads\brno-university-of-technology-ecg-quality-database-but-qdb-1.0.0\brno-university-of-technology-ecg-quality-database-but-qdb-1.0.0\126001\126001_ECG.dat"
hea_path = r"C:\Users\kvjkv\Downloads\brno-university-of-technology-ecg-quality-database-but-qdb-1.0.0\brno-university-of-technology-ecg-quality-database-but-qdb-1.0.0\126001\126001_ECG.hea"

# Read the .dat file
data = np.fromfile(dat_path, dtype=np.float32)

# Print the size of the data
print(f"Size of the data: {data.size}")



# # Visualize a portion of the data
# plt.figure(figsize=(12, 6))
# plt.plot(data[100000:101000])  # Plot a portion of the data
# plt.title("ECG Data Visualization")
# plt.xlabel("Sample Index")
# plt.ylabel("Amplitude")
# plt.grid()
# plt.show()

# Frequency analysis using Welch's method
fs = 500  # Assuming a sampling frequency of 500 Hz
f, psd = welch(data, fs, nperseg=1024)

# Plot the Power Spectral Density (PSD)
plt.figure(figsize=(12, 6))
plt.semilogy(f, psd)
plt.title("Power Spectral Density (PSD) of ECG Data")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Power/Frequency (dB/Hz)")
plt.grid()
plt.show()
```
